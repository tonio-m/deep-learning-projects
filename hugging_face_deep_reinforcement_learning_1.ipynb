```python
# imports
import imageio
import moviepy.editor
import gymnasium as gym
from stable_baselines3 import PPO
from stable_baselines3.common.evaluation import evaluate_policy
```

```python
# example of random run
env = gym.make("LunarLander-v2",render_mode="rgb_array")
env.reset()
writer = imageio.get_writer("file.mp4", fps=30)

for step in range(200):
    frame = env.render()
    writer.append_data(frame)
    env.step(env.action_space.sample())

writer.close()
env.close()
moviepy.editor.ipython_display("file.mp4")
```

```python
env = gym.make("LunarLander-v2", render_mode="rgb_array")
model = PPO("MlpPolicy", env, verbose=1)
model.learn(total_timesteps=int(10e6), progress_bar=True)
# explanation of the logs: https://stable-baselines3.readthedocs.io/en/master/common/logger.html#explanation-of-logger-output
model.save("ppo_lunar")
```

```python
model = PPO.load("ppo_lunar", env=env)
mean_reward, std_reward = evaluate_policy(model, model.get_env(), n_eval_episodes=10)

vec_env = model.get_env()
obs = vec_env.reset()
writer = imageio.get_writer("file.mp4", fps=30)
for i in range(1000):
    action, _states = model.predict(obs, deterministic=True)
    obs, rewards, dones, info = vec_env.step(action)
    frame = vec_env.render()
    writer.append_data(frame)
    if all(dones):
      break
writer.close()
vec_env.close()
moviepy.editor.ipython_display("file.mp4")
```
